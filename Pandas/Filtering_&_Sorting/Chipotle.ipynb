### Step 4. How many products cost more than $10.00?

```python
#this should work, error ocurring maybe because of data type.. item_price is an object, change type to float
#You have to convert the data type to numeric
chipo['new_price']=pd.to_numeric(chipo['item_price'].replace('[$]','',regex=True))
```

```python
chipo.head()
```

```python
#the question asks how many products not orders, so we have to find all unique products in the item_name column
#filter then groupby?
chipo.loc[chipo['new_price']>10.00]
```

### Step 6. Sort by the name of the item

```python
chipo_sorted = chipo.sort_values(by='item_name', ascending=False)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
#sort the data frame in descending order by new_price
chipo.sort_values(by='new_price',ascending=False)
#the quantity is 15
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo.groupby('item_name')['quantity'].sum()
#the Veggie salad bowl was ordered 18 times
```

**EDA using Visualizations**

```python
data['sale_year'] = data['saledate'].str.split(expand=True)[3]
data['sale_month'] = data['saledate'].str.split(expand=True)[1]
```

```python
data['sale_year'] = data['sale_year'].fillna(0)
data['year'] = data['year'].fillna(0)
```

```python
data['Age at Sale'] = data['sale_year'].astype('int') - data['year'].astype('int64')
```

```python
data['year'] = pd.to_datetime(data['year'],format='%Y')
```

```python
data['parsed_date'] = pd.to_datetime(data['saledate'], utc=True, errors='coerce')
```

```python
data.head()
```

```python
#trend of selling price over years
plt.figure(figsize=(14, 5))
sns.lineplot(x='year',y= 'sellingprice', data=data,label='Selling price')
sns.lineplot(x='year',y= 'mmr', data=data,label='mmr')
plt.title('Selling Price Trend')
plt.ylabel('Wams')
plt.legend()

plt.show()
```

```python
correlation=data[['condition','odometer','mmr','sellingprice','Age at Sale']].corr()
```

```python
sns.heatmap(correlation,annot=True)
```

```python
target = data['make'].value_counts().nlargest(3).index
```

```python
target
```

```python
filt_df = data[data['make'].isin(target)]
```

```python
filt_df
```

```python
sns.scatterplot(data=filt_df,x='odometer',y='sellingprice',hue='make')
plt.show()
```

```python
sns.scatterplot(data=filt_df,x='odometer',y='mmr',hue='make')
plt.show()
```

```python
sns.scatterplot(data=filt_df,x='sellingprice',y='mmr',hue='make')
plt.show()
```

```python
sns.kdeplot(x='sellingprice',data=filt_df,hue='make',fill=True)
```

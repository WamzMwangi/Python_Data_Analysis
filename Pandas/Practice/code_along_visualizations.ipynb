**EDA using Visualizations**

```python
data['sale_year'] = data['saledate'].str.split(expand=True)[3]
data['sale_month'] = data['saledate'].str.split(expand=True)[1]
```

```python
data.head()
```

```python
data['sale_year'] = data['sale_year'].fillna(0)
data['year'] = data['year'].fillna(0)
```

```python
data['Age at Sale'] = data['sale_year'].astype('int') - data['year'].astype('int64')
```

```python
data['year'] = pd.to_datetime(data['year'],format='%Y')
```

```python
data['parsed_date'] = pd.to_datetime(data['saledate'], utc=True, errors='coerce')
```

```python
data.head()
```

```python
#trend of selling price over years
plt.figure(figsize=(5,5))
sns.lineplot(x='year',y= 'sellingprice', data=data,label='Selling price')
sns.lineplot(x='year',y= 'mmr', data=data,label='mmr')
plt.title('Selling Price Trend')
plt.ylabel=('Wams')
plt.legend()

plt.show()
```

#### Trend of car prices for the top ten brands...

```python
top_ten=data['make'].value_counts().nlargest(10).index
top_ten
```

```python
top_ten_cars=data[data['make'].isin(top_ten)]
```

```python
top_ten_cars.head()
```

```python
plt.figure(figsize=(12,5))
sns.lineplot(data=top_ten_cars, x='year',y='sellingprice',hue='make')
plt.title('Selling Price Trends of Top 10 Car Brands')
plt.ylabel('Selling Price in USD')
plt.xlabel('YEAR')
plt.legend(loc='lower center', ncol=5)
```

```python
plt.figure(figsize=(12,5))
sns.countplot(data=top_ten_cars, x='make')
```

```python
car_popularity=data['make'].value_counts().reset_index()
car_popularity
```

```python
plt.figure(figsize=(20,5))
sns.barplot(data=car_popularity,x='make',y='count')
plt.xticks(rotation=90)
#plt.ylabel('Car Brand')
#plt.xlabel('Units Sold')

plt.show()
```
